```python
# ---- STEP 1: Create a new clean hourly timestamp index ----
start_date = '2018-04-01 00:00:00'    # adjust if needed
num_rows = len(df)                    # total rows â†’ total hours

new_index = pd.date_range(start=start_date, periods=num_rows, freq='h')

# Assign index to dataframe
df = df.set_index(new_index)
df.index.name = 'Date_Time_Hour_Beginning'

# Remove old datetime column if it still exists
if 'Date Time Hour Beginning' in df.columns:
    df = df.drop(columns=['Date Time Hour Beginning'])

# Clean column names (replace spaces with underscores)
df.columns = df.columns.str.replace(' ', '_')

# Sort index and remove duplicates (safety)
df = df.sort_index()
df = df[~df.index.duplicated(keep='first')]

print("New shape:", df.shape)
df.head()
```

```python
# === STEP 1B: CREATE TIME-BASED FEATURES ===

# Ensure the index is datetime
df.index = pd.to_datetime(df.index)

# Create time-based features
df['hour'] = df.index.hour
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df['dayofweek'] = df.index.dayofweek          # Monday=0, Sunday=6
df['is_weekend'] = (df['dayofweek'] >= 5).astype(int)

# Month/Year boundary features
df['is_month_start'] = df.index.is_month_start.astype(int)
df['is_month_end'] = df.index.is_month_end.astype(int)
df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
df['is_quarter_end'] = df.index.is_quarter_end.astype(int)

# Hour of Year (optional but useful)
df['hour_of_year'] = df.index.dayofyear * 24 + df.index.hour

# Display confirmation
df[['hour','day','month','year','dayofweek','is_weekend','is_month_start','is_month_end']].head()
```

```python
import numpy as np
import pandas as pd

# Assuming your dataframe is df and your target column is target_col
# 'month' and 'hour' columns must already exist
# Example target column (change if needed)


# Function for monthly-hourly median imputation
def impute_month_hour(col_name):
    return df[col_name].fillna(
        df.groupby(["month", "hour"])[col_name].transform("median")
    )

# List all numeric columns to impute
numeric_cols_to_impute = df.select_dtypes(include=np.number).columns.tolist()

# Apply 3-step imputation + final median fallback
for col_to_impute in numeric_cols_to_impute:
    if col_to_impute != target_col:
        # Step 1: Monthly-Hourly Median
        df[col_to_impute] = impute_month_hour(col_to_impute)

        # Step 2: Forward fill
        df[col_to_impute] = df[col_to_impute].ffill()

        # Step 3: Backward fill
        df[col_to_impute] = df[col_to_impute].bfill()

        # Step 4: Fill any remaining NaNs with overall median
        df[col_to_impute] = df[col_to_impute].fillna(df[col_to_impute].median())

# Optionally, do the same for the target column if you want no missing values
# df[target_col] = df[target_col].fillna(df[target_col].median())

# Check all numeric columns
print(df[numeric_cols_to_impute].isna().sum())
```

```python
# PACF-informed lags
lags = [1, 2, 3, 6, 12, 24]
for lag in lags:
    df[f"lag_{lag}"] = df["Residual_Demand"].shift(lag)

# Rolling windows
df["roll3"] = df["Residual_Demand"].rolling(3).mean()
df["roll6"] = df["Residual_Demand"].rolling(6).mean()
df["roll12"] = df["Residual_Demand"].rolling(12).mean()
df["roll24"] = df["Residual_Demand"].rolling(24).mean()

# Time features
df["hour"] = df.index.hour
df["dayofweek"] = df.index.dayofweek
df["month"] = df.index.month

# Drop rows with NaN created by lags
df = df.dropna()
```

TIME-BASED TRAIN/TEST SPLIT

```python
# STEP 5 â€” Train/Test Split (time-based)

# Use the last 5000 rows as the test set (you can adjust)
test_size = 5000

train = df.iloc[:-test_size].copy()
test = df.iloc[-test_size:].copy()

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# Target variable
y_train = train["Residual_Demand"]
y_test = test["Residual_Demand"]

# Features (everything except target)
feature_cols = [c for c in df.columns if c != "Residual_Demand"]

X_train = train[feature_cols]
X_test = test[feature_cols]

print("\nFeature set prepared:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

# Train LightGBM on X_train / y_train

### *Comparision of models *
LightGBM: Excellent â€“ Best overall performance (especially for electricty demand).

Linear Regression : Poor â€” Electricity demand is not linear.

Random Forest: Moderate â€” Too slow for large feature set.

XGBoost: Very Good â€” But slower and heavier.



```python
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Datasets
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test)

# Parameters
params = {
    "objective": "regression",
    "metric": "rmse",
    "learning_rate": 0.03,
    "num_leaves": 64,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "verbose": -1,
}

# --- FIX: early stopping via callbacks ---
callbacks = [
    lgb.early_stopping(stopping_rounds=200),
    lgb.log_evaluation(period=200)
]

# Train
model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, test_data],
    num_boost_round=5000,
    callbacks=callbacks
)

# Predict
y_pred = model.predict(X_test, num_iteration=model.best_iteration)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print("Performance on Test Set:")
print("RMSE:", rmse)
print("MAE:", mae)
```

# Computing key error metrics

1. MAE â€” Mean Absolute Error: MAE is the average absolute difference between predicted and actual values.

2. RMSE â€” Root Mean Squared Error:RMSE tells you how far the modelâ€™s predictions are from the actual values, giving extra penalty to large errors.

3. MAPE â€” Mean Absolute Percentage Error: MAPE shows the average percent error, telling you how accurate the model is on a percentage basis.

4. RÂ² â€” Coefficient of Determination:RÂ² tells you how much of the variation in the data the model is able to explain.

## **1. Purpose of the Bias Review**

The goal of this bias analysis is to identify, evaluate, and mitigate systematic errors in the LightGBM demand forecasting model that could affect operational decision-making, resource allocation, and reliability planning at Eskom.

---

## **2. Sources of Potential Bias**

Bias in time-series energy forecasting typically arises from *data irregularities*, *seasonal distortions*, and *external disruptions*. For this project, the following sources were evaluated:

### **2.1 Data Quality Bias**

* Missing values from telemetry gaps or meter issues.
* Inconsistent timestamps due to daylight savings, logging shifts, or format errors.
* Historical anomalies caused by temporary grid instability.

### **2.2 Load-Shedding Distortion Bias**

Demand artificially drops during load-sheddingâ€”not because South Africans use less electricity, but because supply is constrained.
If these periods are not handled, the model can *under-learn* true peak demand behavior.

### **2.3 Seasonal Bias**

* Winter months show sharp morning/evening peaks.
* In summer, the pattern flattens slightly.

If the model trains disproportionately on one season, predictions become skewed.

### **2.4 Holiday & Event Bias**

Public holidays, long weekends, and national events create atypical consumption patterns.
Models may misinterpret these as random noise or outliers.

### **2.5 Time-of-Day Variance Bias**

Higher natural variance during peak hours (morning + evening) increases forecasting difficulty and may bias the model toward under-predicting peak consumption.

---

## **3. Bias Patterns Observed During Model Evaluation**

### **3.1 Under-Prediction During Winter Peaks**

Residual analysis indicates the model consistently predicts slightly lower than true observed values during high-demand winter mornings/evenings.

### **3.2 Over-Prediction During Load-Shedding Events**

Because demand drops artificially, the model sometimes forecasts higher than the constrained demand recorded.

### **3.3 Misinterpretation of Holiday Effects**

Certain holiday periods were flagged as outliers, indicating the model struggled to generalize these infrequent patterns.

### **3.4 Higher Error During Volatile Periods**

RMSE spikes during:

* Sudden heatwaves
* Generator trips affecting supply
* Unexpected consumption surges

These are known system volatility periods.

---

## **4. Mitigation Strategies Implemented**

### **4.1 Seasonal Feature Engineering**

Added:

* Month
* Season
* Day-of-week
* Hour of day
* Business-day indicator
  â€¦to reduce temporal bias.

### **4.2 Load-Shedding Indicator Feature**

Introduced a binary feature marking load-shedding hours, preventing the model from learning those values as normal demand.

### **4.3 Monthly-Hourly Median Imputation**

This method retains natural seasonal structure while avoiding biased spikes from missing values.

### **4.4 Outlier Smoothing for Abnormal Dips/Spikes**

Applied rolling-window smoothing for extreme points not aligned with regular consumption behavior.

### **4.5 Time-Based Cross-Validation**

Ensured no look-ahead bias by splitting data sequentially, reducing overly optimistic error estimates.

---

## **5. Real-World Implications of Bias**

### **Negative Impacts If Unaddressed**

* **Underestimating peak demand** leads to insufficient generator scheduling â†’ higher risk of system stress.
* **Over-forecasting during load-shedding** creates false signals about available elasticity.
* **Poor holiday forecasting** could misguide operational planning for special events and long weekends.
* **Seasonal under-performance** affects policy-level decisions on reserve capacity and tariffs.

### **Positive Outcomes After Mitigation**

* Improved generation dispatch accuracy.
* Reduced operational costs due to better peak planning.
* More reliable short-term forecasts for control room operations.
* Better alignment with national energy planning goals.
